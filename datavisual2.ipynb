```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set(style="white", color_codes=True)
df=pd.read_csv('finaldatasetnew.csv')
fig, ax = plt.subplots(figsize=(15,12))
y_pos = np.arange(len(df['Country']))
df.sort_values('Economy (GDP per Capita)', ascending=True, inplace = True)
ax.barh(y_pos, df['Economy (GDP per Capita)'], align='center', color='green', ecolor='black');
ax.set_yticks(y_pos)
ax.set_yticklabels(df['Country'],fontsize=14);
  # labels read top-to-bottom
ax.set_xlabel(df.columns.values.tolist()[15],fontsize=18);
ax.set_xticklabels(df['Economy (GDP per Capita)'],fontsize=14);
ax.set_xlabel('GDP');
ax.set_title('ECONOMIES OF HAPPY COUNTRIES',fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
sns.set(style="white", color_codes=True)
df=pd.read_csv('finaldatasetnew.csv')
plt.subplots(figsize=(20,16))
x=df['Country']
y=df['Health (Life Expectancy)']
sns.barplot(x, y, hue='Region', data=df, saturation=0.8)
plt.grid(False)
plt.xticks(rotation=90)
plt.xticks(rotation=90,fontsize=18)
plt.yticks(rotation=90,fontsize=18)
plt.xlabel('Country', fontsize=18)
plt.ylabel('Health (Life Expectancy)', fontsize=18) 
plt.title('COUNTRY vs LIFE EXPECTANCY',fontsize=22)
sns.set_style({'figure.facecolor': 'white'})
plt.show()
```

```python
w_country=df['Region'].value_counts()
top_10=w_country[:25]
plt.figure(figsize=(15,9))
sns.barplot(top_10.index,top_10.values)
plt.xticks(rotation=90,fontsize=14)
plt.yticks(fontsize=14)
plt.xlabel('REGION',size=14,labelpad=20,fontsize=18)
plt.ylabel('NUMBER OF COUNTRIES',size=14,labelpad=20,fontsize=18)
plt.title('HAPPIEST REGION',size=18)
plt.grid(False)
plt.show()
```
